```python
import pandas as pd
import numpy as np
import scipy as sp
import sklearn as sk
import matplotlib.pyplot as plt
# import the date module
from datetime import date
```

```python
# Store filepath in a variable: 
#co2_1 = "co2.csv"
co2_1 = "CO2_bynation_1751_2014.csv"
co2_1_df = pd.read_csv(co2_1)
#co2_1_df = pd.read_csv(co2_1,parse_dates=['year'])
co2_1_df.info()
co2_1_df.head()
```

```python
# Rename the data columns
co2_r_df=co2_1_df.rename(columns={'Nation':'country',
                             'year': 'year', 
                              'co2' :'co2' })
co2_r_df
```

```python
# drop the NaN values and note the details of the dropped data.
co2_r_df= co2_r_df.dropna(axis=0)
co2_r_df.describe()
co2_r_df.info()
```

```python
#determine   what  is the date  range of df  at this step

co21_min_date = co2_r_df["year"].min()
co21_max_date = co2_r_df["year"].max()
print("min date " + str(co21_min_date) + " max date " + str(co21_max_date))
```
