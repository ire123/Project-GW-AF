```python
#temp_corr = "clean_GlobalLandTemperaturesByCountry_co2.csv"
temp_corr = "clean_YearLandAvgTempsByCountry.csv"
temp_corr_df = pd.read_csv(temp_corr, encoding="cp1252")
temp_corr_df.describe()
temp_corr_df.info()
temp_corr_df.head()
```

```python
#temperature DF:  group by year and get annual mean
#use for  correlation factor
temp_mean_df=(temp_corr_df.groupby(["year"]).agg({'avg_yly_tmp_f':'mean'}))
temp_mean_df
```

```python
#this is for correlation factor to prove  data  is sound and valid.  Not widely skewed up and down form year to year
temp_mean_df.corr()
```

```python
#temperature DF:  group by country, year and get annual mean.  use for linear and other plots
#showing correlation factor
temp_mean_ctry_df=(temp_corr_df.groupby(['year', 'country']).agg({'avg_yly_tmp_f':'mean'}))

temp_mean_ctry_df.head()
```

```python
temp_mean_ctry_df.corr()
```

```python
# Lag Plot of temp data
# Lag plots are used to check if a data set or time series is random.
# Random data should not exhibit any structure in the lag plot. 
# Non-random structure implies that the underlying data are not random.

from pandas.plotting import lag_plot
plt.figure()


data = pd.Series(0.1 * np.random.rand(1000) +
     0.9 * np.sin(np.linspace(-99 * np.pi, 99 * np.pi, num=1000)))
lag_plot(data)
plt.show()
```
