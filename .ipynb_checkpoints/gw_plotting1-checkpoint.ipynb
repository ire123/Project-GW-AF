```python
import pandas as pd
import numpy as np
import scipy as sp
```

```python
# Store filepath in a variable: use parse_dates=["date"] 
# 
gltc = "./data/clean/clean_YearLandAvgTempsByCountry.csv"

# Read our Data file with the pandas library
gltc_df = pd.read_csv(gltc,encoding="ISO-8859-1")
#gltc_df.describe()
#gltc_df.info()
gltc_df.head()
```

```python
# Set the index to the date column for pandas df.plot()
gltc1_df = gltc_df.loc[gltc_df["Country"] == "United States"]
#print(len(gltc1_df.index))
#gltc1_df.info()
gltc1_df = gltc1_df.sort_values(["year"], ascending=[True]).\
                                reset_index(drop=True)
gltc1_df.head(5)
```

```python
# Dependencies
from matplotlib.figure import Figure
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime as dt
from datetime import datetime, date, time
import matplotlib.dates as mdates

fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(111)

x_axis = gltc1_df["year"]  
y_axis = gltc1_df["avg_yly_tmp_f"]

ax.set_title("USA Average Temperatures 1768 - 2013", weight='bold').set_fontsize('18')
ax.set_xlabel("January 1768 - September 2013").set_fontsize('14')
ax.set_ylabel("Average Temperatures(F)").set_fontsize('14')

ax.grid(True)
ax.set_facecolor("whitesmoke")
# fig.autofmt_xdate(rotation=45, ha='right')
# ax.xaxis.set_major_locator(mdates.YearLocator())
# ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

#ax.plot(x_axis, y_axis, color="magenta", linewidth=1)
ax.plot(x_axis, y_axis, color="magenta", linewidth=.5)

ax.legend(["USA Avg Temp (F)"], loc="upper left")

plt.tight_layout()
plt.savefig("1768-2013_usa_temp_chart.png")
plt.show()
```

```python
# Set the index to the date column for pandas df.plot()
gltc2_df = gltc_df.loc[gltc_df["Country"] == "China"]
print(len(gltc2_df.index))
#gltc2_df.info()
gltc2_df = gltc2_df.sort_values(["year"], ascending=[True]).\
                                reset_index(drop=True)
gltc2_df.head(5)
```

```python
# Get the min and max date range
gltc2_min_date = gltc2_df["year"].min()
gltc2_max_date = gltc2_df["year"].max()
print("min date " + str(gltc2_min_date) + " max date " + \
      str(gltc2_max_date))
```
